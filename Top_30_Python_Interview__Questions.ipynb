### Q1 ) Write a Python program to reverse a string without using (.reverse() function) .
How many different ways do you know?

```python
# 4 Ways (One-Linear , using a loop, Custom Func, List Comp )
```

```python
# Method 1 : one linear Pythonic Way!!

word = 'python'
rev_word = word[::-1]
print(rev_word)
```

```python
# method 2 : Using loop

word = "Python"

result = ''
for char in word:
  result = char + result  #### char is adding before rsult!!!! ---> ....... y + P

print(result)
```

```python
# Method 3 : Using Custom Functions!!

def reverse_word(word):

  result=""

  for char in word:
    result = char + result

  return result


word1 = "Data Science"
print(reverse_word(word1))
```

```python
# method 4 : List Comprehension !!!   ### .reverse() allowed here!

word='Prajwal'

rev_word = "".join([char for char in reversed(word)])

print(rev_word)
```

### Q2) Find the Second Highest number in the list .

```python
# 3 Ways (using sort() , using max() function , without built- in functions)
```

```python
# Method 1 : Using Sort() Function

lst=[23,44,75,87,53,29,69]

lst.sort(reverse=True) ####

print(lst[1]) ## used indexing
```

```python
lst=[23,44,75,87,53,29,69]

lst.sort()
print(lst[-2])
```

```python
# Method 2 : Using Max() function

lst=[23,44,75,87,53,29,69]

largest = max(lst)

l = [i for i in lst if i != largest]

largest_2nd = max(l)

print(largest_2nd)
```

```python
# Method 3 : without using any Built-in Function

def second_largest(lst):

  largest = float("-inf")
  largest_2nd =float("-inf")

  for i in lst:
    if i > largest :
      largest_2nd = largest  ######
      largest =i          ###

    elif i > largest_2nd and i != largest : ##
      largest_2nd = i

  return largest_2nd
```

```python
lst1=[23,44,75,87,53,29,69]
print(second_largest(lst1))
```

### Q3) Flatten a Nested List Using a Single List Comprehension .

nested_list = [[1,2,3],[4,5],[6,7,8,9]]

*#expected output :* [1,2,3,4,5,6,7,8,9]

```python
# 2 Ways : (without list comprehension , With list Comprehension)
```

```python
#Flatten the Nested list


# Method 1 : List Comprehension

nested_list = [[1,2,3],[4,5],[6,7,8,9]]

new_list = [ j for i in nested_list for j in i ]   ####

print(new_list)
```

```python
# Method 2 : Without Nested Loop

nested_list = [[1,2,3],[4,5],[6,7,8,9]]

new_list=[]

for i in nested_list:
  for j in i:
    new_list.append(j)

print(new_list)
```

```python
### condition ###

nested_list = [[1,2,3],[4,5],11,12,[6,7,8,9]]

new_list = []

for i in nested_list:
  if type(i) == list:
    for j in i:
      new_list.append(j)

  elif type(i) != list:    ### checks the type of value in the list!!! ##11,12
    new_list.append(i)

print(f'The new flatten list : {new_list}')
```

### Q5) Move zeros to End : Given an array of integers,move all 0s to teh end of it while maintaining the relative order of the non-zero elements

nums = [0,1,03,7,8,0,12]

```python
### Method 1

nums = [0,1,0,3,7,8,0,12]

zeros = []
inn =[]

for i in nums :
  if i == 0:
    zeros.append(i)

  else:
    inn.append(i)

print(zeros)
print(inn)
print(inn + zeros)
```

```python
### Method 2

nums = [0,1,0,3,7,8,0,12]

lst_non_zeros = [i for i in nums if i != 0]
print(f'non-zero_nums : {lst_non_zeros}')

lst_zeros = [0] * (len(nums) - len(lst_non_zeros))
print(f'zero-nums : {lst_zeros}')

new_lst = lst_non_zeros + lst_zeros
print(new_lst)
```

### Q6) Vowel/Consonant Counter : Create a function that accurately counts vowels and consonants in a string using list comprehension.

```python
def vowel_consonant_count(text):

  vowels = 'aeiou'

  var_ = [i for i in text.lower() if i in vowels]
  vowel = len(var_)

  con_ =[i for i in text.lower() if i.isalpha() and i not in vowels]
  consonent = len(con_)

  return vowel,consonent

v,c = vowel_consonant_count('Hello World Python 1234')
print(f'The vowels are {v} and the consonents are {c} in the text')

```

```python
v,c = vowel_consonant_count('I am Prajwal, Working as ML Engineer as Uniathena')
print(f'The vowels are {v} and the consonents are {c} in the text..!')
```

### Q7) Compress a string by counting consecutive character
Input : "aaabbbc" , "aabbcaab"  | Output : "a3b2c1" , "a2b2c1a2b1  

```python
## Method 1 :

# Input : "aaabbc"  | Output : "a3b2c1"

text= "aaabbc"

result=""
count=1

for i in range(1,len(text)) :
  if text[i] == text[i-1]:
    count +=1

  else:
    result += text[i-1] + str(count)  # all variables
    count=1

result += text[-1] + str(count) # For last variable
print(result)
```

### Q8) Extract Prime Numbers from List of integers from 1 to 100 .
Using filter() and a lambda function.

```python
# what are prime numbers ?
# the value that divide by 1 or by itself.
#eg. 11--> 1,11 can only divide (11). ; 14--> not a prime no. as it can be divided by 2,7,14 !
```

```python
numbers = list(range(1,101))
print(numbers)
```

```python
# method 1

primes = list(filter(lambda x : x > 1 and all(x % i != 0 for i in range(2,x)), numbers))

print(primes)

```

```python
primes_lst=[]

for x in numbers:
  if x > 1 :  # prime_no starts from 2
    is_prime = True

    for i in range(2,x):
      if x % i == 0 :
        is_prime = False
        break

    if is_prime:
      primes_lst.append(x)

print(primes_lst)

```

```python
# method 2

numbers = list(range(1,101))
print(numbers)

primes_list=[]

for x in numbers:
  if x > 1:
    for i in range(2,x):
      if x % i == 0:
        break

    else:
      primes_list.append(x)

print(f'Prime no. (1-100) : {primes_list}')
```
